```python
# cpu faiss with intel MLK lib use BLAS Intel10_64lp model and openMP lib run fast
in_colab = 'google.colab' in str(get_ipython())
if in_colab:
  !source /opt/intel/oneapi/mkl/latest/env/vars.sh && echo $MKLROOT && echo $CPATH \
   && cd faiss && rm -rf build && cmake -B build \
      -DBUILD_SHARED_LIBS=ON \
      -DFAISS_ENABLE_C_API=ON \
      -DBUILD_TESTING=ON \
      -DFAISS_ENABLE_GPU=OFF \
      -DFAISS_ENABLE_RAFT=OFF \
      -DFAISS_ENABLE_PYTHON=OFF \
      -DBLA_VENDOR=Intel10_64lp \
      -DFAISS_OPT_LEVEL=avx2 \
      -DCMAKE_INSTALL_LIBDIR=lib \
      -DCMAKE_BUILD_TYPE=Release .
```

```python
!cd faiss && make -C build -j faiss faiss_avx2
```
